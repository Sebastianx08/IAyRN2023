```python
#Escribe un programa que tome una lista de números ingresados por el usuario y muestre la lista en orden inverso.

numero = input("Introduce una lista de numeros: ")
lista = [int(numero) for numero in numero.split(',')]
lista.reverse()
print("La lista inversa es: ",lista)
```

```python
"""
Juego de adivinar el número: Crea un programa que genere un número aleatorio entre 1 y 100, y luego le pida al usuario que adivine el número.
El programa debe proporcionar pistas al usuario si el número es mayor o menor que el número
objetivo, y seguir solicitando un nuevo intento hasta que el usuario adivine correctamente.
"""

import random
num_aleatorio = random.randint(1,100)
while True:
  usuario = int(input("Adivina un numero entre 1 y el 100: "))
  if usuario > num_aleatorio:
    print("Upss!!, te pasaste!, INTENTA CON UN NUMERO MAS PEQUENO")
  elif usuario < num_aleatorio:
    print("Upss!!, es muy bajo, INTENTA CON UN NUMERO MAS GRANDE")
  else:
    print("FELICIDADES, HAS ADIVINADO EL NUMERO!!")
    break
```

```python
""" Ejercicio de cálculo de números primos en un rango dado: Escribe un programa que
    solicite al usuario un rango de números y muestre todos los números primos dentro de ese rango.
"""

def primo(numero_primo):
    if numero_primo < 2:
      return False
    for i in range(2, int(numero_primo**0.5) + 1):
      if numero_primo % i == 00:
        return False
    return True
inicio = int(input("Ingrese el inicio del rango: "))
fin    = int(input("Ingrese el fin del rango: "))
print("Los numero primos en el rango son: ")
for numero in range(inicio, fin + 1):
  if primo(numero):
    print(numero)
```
